# TS Diagrams

Modified from https://github.com/COSIMA/cosima-recipes/blob/main/DocumentedExamples/TemperatureSalinityDiagrams_mom5_mom6.ipynb

```python

box1latslice = slice(box1latS,box1latN)
box1lonslice = slice(box1lonW,box1lonE)
box2latslice = slice(box2latS,box2latN)
box2lonslice = slice(box2lonW,box2lonE)
expt = '01deg_jra55v140_iaf_cycle4'
start_time = '2000-01-01'
end_time = '2018-12-31'
timeslice = slice(start_time,end_time)
```

```python
def gsw_SA_from_SP(salt):
    """function to convert practical salinity to absolute salinity
    using the Gibbs SeaWater (GSW) Oceanographic Toolbox of TEOS-10
    (https://teos-10.github.io/GSW-Python/)
    
    input:
        -salt: practical salinity (array)
        -lon_name: name of the longitude (str)
    output:
        -salt_abs: absolute salinity (array)
    """
    pres = gsw.p_from_z(-salt.st_ocean, salt.yt_ocean)
    salt_abs = gsw.SA_from_SP(salt, pres, salt.xt_ocean, salt.yt_ocean)
    salt_abs.attrs = {'units': 'Absolute Salinity (g/kg)'}
    return salt_abs
```

```python
def gsw_CT_freezing(salt):
    CT_freezing = gsw.CT_freezing(salt, 10.1325,0)
    return CT_freezing
```

```python
# a function to load area of grid cells
def load_grid_areas(lon,lat):
    area = cc.querying.getvar(variable='area_t', session=session, n=1, expt = expt)
    #area = area.drop(['geolon_t', 'geolat_t'])
    # select longitude and latitude range
    area = area.sel(xt_ocean = lon, yt_ocean = lat)
    return area
```

```python
# a function that computes the temperature and salinity bins for 2D histogram 
def compute_TS_bins(salt, temp, area):
    temp_bins = np.arange(np.floor(temp.min().values), np.ceil(temp.max().values), 0.05)
    salt_bins = np.arange(np.floor(salt.min().values), np.ceil(salt.max().values), 0.1)
    # for density contours in TS diagram
    temp_bins_mesh,salt_bins_mesh = np.meshgrid(temp_bins, salt_bins)
    TS_density = gsw.density.sigma2(salt_bins_mesh, temp_bins_mesh)
    # volume of grid cells to account for varying grid cells especially in the vertical
    vol = (temp.st_ocean * area)

    # 2D histogram of temperature and salinity weighted by volume
    TS_hist = xhistogram(
        temp, salt, bins=(temp_bins, salt_bins), weights=vol)
    TS_hist = TS_hist.where(TS_hist != 0).compute()
    return TS_hist, TS_density, salt_bins_mesh, temp_bins_mesh 
```

```python
fig, axes = plt.subplots(ncols = 2, figsize=(15, 8))

lat = box1latslice
lon = box1lonslice

temp = cc.querying.getvar(
        variable='temp', session=session, expt = expt, 
        start_time = start_time, end_time = end_time).sel(time = timeslice, xt_ocean = lon, yt_ocean = lat)-273.15
salt = cc.querying.getvar(
        variable='salt', session=session, expt = '01deg_jra55v140_iaf_cycle4', 
        start_time = start_time, end_time = end_time).sel(time = timeslice, xt_ocean = lon, yt_ocean = lat)

salt = gsw_SA_from_SP(salt)
area = load_grid_areas(lon,lat)
TS_hist, TS_density, salt_bins_mesh, temp_bins_mesh = compute_TS_bins(salt, temp, area)

norm=colors.LogNorm(
    vmin=TS_hist.min().values, vmax=TS_hist.max().values)
# volume weighted histogram of T and S
TS_hist.plot(ax = axes[0], cmap=cmo.deep, norm=norm,
             cbar_kwargs=dict(label='volume ($m^{3}$)'))
# density contours
cs = axes[0].contour(
    salt_bins_mesh, temp_bins_mesh, TS_density, colors='silver',
                 levels=np.arange(np.floor(TS_density.min()),
                                  np.ceil(TS_density.max()), .5))
axes[0].clabel(cs, inline=True)

axes[0].set_xlabel('Absolute salinity (g/kg)')
axes[0].set_ylabel('Conservative temperature ($^\circ$C)')
axes[0].set_title('Box 1 T-S all depths')

lat = box2latslice
lon = box2lonslice

temp = cc.querying.getvar(
        variable='temp', session=session, expt = expt, 
        start_time = start_time, end_time = end_time).sel(time = timeslice, xt_ocean = lon, yt_ocean = lat)-273.15
salt = cc.querying.getvar(
        variable='salt', session=session, expt = '01deg_jra55v140_iaf_cycle4', 
        start_time = start_time, end_time = end_time).sel(time = timeslice, xt_ocean = lon, yt_ocean = lat)

salt = gsw_SA_from_SP(salt)
area = load_grid_areas(lon,lat)
TS_hist, TS_density, salt_bins_mesh, temp_bins_mesh = compute_TS_bins(salt, temp, area)

norm=colors.LogNorm(
    vmin=TS_hist.min().values, vmax=TS_hist.max().values)
# volume weighted histogram of T and S
TS_hist.plot(ax = axes[1], cmap=cmo.deep, norm=norm,
             cbar_kwargs=dict(label='volume ($m^{3}$)'))
# density contours
cs = axes[1].contour(
    salt_bins_mesh, temp_bins_mesh, TS_density, colors='silver',
                 levels=np.arange(np.floor(TS_density.min()),
                                  np.ceil(TS_density.max()), .5))
axes[1].clabel(cs, inline=True)

axes[1].set_xlabel('Absolute salinity (g/kg)')
axes[1].set_ylabel('Conservative temperature ($^\circ$C)')
axes[1].set_title('Box 2 T-S all depths')

#plt.ylabel(temp.units)
#plt.title("T-S with "+model+" ouput");
```
